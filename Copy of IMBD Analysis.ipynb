```python
# How has the distribution of movie genres changed
# over the years?

# Splitting the 'Genres' into separate rows
expanded_frame = df_copy.assign(Genre=df_copy['Genre'].str.split(', ')).explode('Genre')

# Grouping data by 'Year' and 'Genre', then
# getting count of number of movies for each Genre per year.
genre_counts = expanded_frame.groupby(['Released_Year', 'Genre']).size().reset_index(name='Count')

# Pivoting data to get years as index, genres as columns,
# and counts as values
genre_distribution = genre_counts.pivot(index='Released_Year', columns='Genre', values='Count')

# Filling NA values with 0
genre_distribution.fillna(0, inplace=True)

genre_distribution.head()
```

```python
plt.figure(figsize=(15, 10))

# Get total count of movies for each genre across all years
total_counts = genre_distribution.sum()

# Get top 10 genres
top_genres = total_counts.nlargest(5).index

# Plot a line for each of the top genres
for genre in top_genres:
    plt.plot(genre_distribution.index, genre_distribution[genre], label=genre)

# Set the title and labels
plt.title('Count of Movies for Each Genre Over the Years(Largest 5)')
plt.xlabel('Year')
plt.ylabel('Count')

# Add a legend
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Display the plot
plt.show()

```

```python
# Correlation between 'No_of_Votes' and 'IMDB_Rating'
correlation = df_copy['No_of_Votes'].corr(df_copy['IMDB_Rating'])
fig, ax = plt.subplots(figsize=(15, 10))
ax.scatter(df_copy['No_of_Votes'], df_copy['IMDB_Rating'])
ax.set(title='Correlation between Number of Votes  and IMDb Rating', xlabel='No_of_Votes', ylabel='IMDB_Rating')
plt.show()
print(f"The correlation between the number of votes and IMDb rating is {correlation}")
```

```python
# Which genre has the highest average gross revenue per movie?

# Split the genres and expand them into separate rows
expanded_frame = df_copy.assign(Genre=df_copy['Genre'].str.split(', ')).explode('Genre')

# Calculate the average gross revenue for each genre
average_gross = expanded_frame.groupby('Genre')['Gross'].mean()

# Get the genre with the highest average gross revenue
highest_avg_gross_genre = average_gross.idxmax()

print(f"The genre with the highest average gross revenue per movie is {highest_avg_gross_genre}. Its value is {average_gross[highest_avg_gross_genre]:,.2f} dollars.")
```

```python
# Calculate the average gross revenue for each genre
average_gross = expanded_frame.groupby('Genre')['Gross'].mean()

# Create a bar plot
average_gross.plot(kind='bar', figsize=(15, 10))

# Set the title and labels
plt.title('Average Gross Revenue for Each Genre')
plt.xlabel('Genre')
plt.xticks(rotation=70)
plt.ylabel('Average Gross Revenue')

# Display the plot
plt.show()
```
